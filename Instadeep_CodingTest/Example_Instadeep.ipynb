```python
import numpy as np
import pandas as pd

def calcMissing(readings):
    # Write your code here
    x = []
    y = []
    missing_x = []
    for idx,i in enumerate(readings):
        _,date,high = i.split()
        if high[:4]=='Miss':
            y.append(np.nan)
            missing_x.append(idx)
        else:
            # x.append(idx)
            y.append(float(high))
    # x = np.array(x)
    pd_y = pd.Series(y)
    pd_y = pd_y.interpolate(limit_direction='both')
    pd_y = pd_y.array
    # print(pd_y)
    for i in pd_y[missing_x]:
        print(round(i,2))

if __name__ == '__main__':
    readings_count = int(input().strip())

    readings = []

    for _ in range(readings_count):
        readings_item = input()
        readings.append(readings_item)

    calcMissing(readings)
```

```python
#!/bin/python3

import math
import os
import random
import re
import sys
# Complete the 'predictMissingHumidity' function below.
# The function is expected to return a FLOAT_ARRAY.
# The function accepts following parameters:
#  1. STRING startDate
#  2. STRING endDate
#  3. STRING_ARRAY knownTimestamps
#  4. FLOAT_ARRAY humidity
#  5. STRING_ARRAY timestamps
import numpy as np
import pandas as pd

def predictMissingHumidity(startDate, endDate, knownTimestamps, humidity, timestamps):
    
    # algorithm explanation:
    # insert np.nan into all vacant slots of knownTimestamps
    # and use pandas.interpolate() to estimate
    
    new_humidity = []
    missing_stamps_idx = []
    curr_stamp = 0
    prev_stamp = 23

    # insert timestamps into knownTimestamps
    for timstam,hum in zip(knownTimestamps,humidity):
        
        date, tim = timstam.split()
        # store hour-info: '03:00' --> '03'
        curr_stamp = int(tim[:2])
        diff = curr_stamp - prev_stamp
        # sequantial in [0,23], intervals exist
        if diff>1:
            # if more than 1 hour are missing, then add np.nan for times of (diff-1)
            # e.g.
            # 08:00 11:00 --> insert (11-8)-1 = 2 hours: 09:00 10:00
            for k in range(diff-1):
                new_humidity.append(np.nan)
                missing_stamps_idx.append(len(new_humidity)-1)
            new_humidity.append(hum)
        # normal situation, sequantial in [0,23]
        elif diff == 1:
            new_humidity.append(hum)
        # nor mal situation when hour moves from 23:00 to 00:00 of next day
        elif diff == -23:
            new_humidity.append(hum)
        # intervals exist between days
        elif diff <= 0 and not diff == -23:
            for j in range(diff+23):
                new_humidity.append(np.nan)
                missing_stamps_idx.append(len(new_humidity)-1)
            new_humidity.append(hum)
            
        prev_stamp = curr_stamp
    
    # in case the end of hour is still away from 23:00
    end_diff = len(timestamps) - len(missing_stamps_idx)
    if end_diff > 0:
        for s in range(end_diff):
            new_humidity.append(np.nan)
            missing_stamps_idx.append(len(new_humidity)-1)
            
    
    pd_new_humidity = pd.Series(new_humidity)
    pd_new_humidity = pd_new_humidity.interpolate(limit_direction='both')
    pd_new_humidity = pd_new_humidity.array

    return pd_new_humidity[missing_stamps_idx]
        
    

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    startDate = input()

    endDate = input()

    knownTimestamps_count = int(input().strip())

    knownTimestamps = []

    for _ in range(knownTimestamps_count):
        knownTimestamps_item = input()
        knownTimestamps.append(knownTimestamps_item)

    humidity_count = int(input().strip())

    humidity = []

    for _ in range(humidity_count):
        humidity_item = float(input().strip())
        humidity.append(humidity_item)

    timestamps_count = int(input().strip())

    timestamps = []

    for _ in range(timestamps_count):
        timestamps_item = input()
        timestamps.append(timestamps_item)

    result = predictMissingHumidity(startDate, endDate, knownTimestamps, humidity, timestamps)

    fptr.write('\n'.join(map(str, result)))
    fptr.write('\n')

    fptr.close()
```
